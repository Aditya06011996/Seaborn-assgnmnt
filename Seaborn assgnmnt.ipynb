Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans - Seaborn is a popular Python data visualization library that provides various types of plots for exploring and visualizing data. Here are five common plots that you can create using Seaborn along with their typical use cases:

(1) Histograms:

Use: Histograms are used to visualize the distribution of a single numerical variable. They show the frequency or probability density of data within different bins or intervals.

(2) Scatter Plots:

Use: Scatter plots are used to visualize the relationship between two numerical variables. They help identify patterns, trends, and correlations in the data.

(3) Bar Plots:

Use: Bar plots are used to display categorical data with rectangular bars. They are great for comparing the values of different categories and showing how they relate to each other.

(4) Box Plots (Box-and-Whisker Plots):

Use: Box plots are used to visualize the distribution of a numerical variable and to identify outliers. They display the median, quartiles, and potential outliers in the data.

(5) Heatmaps:

Use: Heatmaps are used to visualize data in a matrix format. They are particularly useful for displaying correlation matrices or highlighting patterns in large datasets.

These are just a few examples of the many plot types available in Seaborn. Depending on your specific data analysis needs, Seaborn offers a wide range of additional plots like line plots, violin plots, pair plots, and more to help you effectively explore and communicate your data.






Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

Ans -To load the "fmri" dataset and plot a line plot using Seaborn for different events and regions, you can follow these steps:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot using Seaborn
plt.figure(figsize=(10, 6))  # Optional: Set the figure size

# Use the sns.lineplot() function to create the line plot
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")

# Set plot labels and title
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("FMRI Signal Over Time for Different Events and Regions")

# Show the legend
plt.legend(title="Event")

# Show the plot
plt.show()
```

In this code:

1.We import the necessary libraries, including Seaborn and Matplotlib.

2.We load the "fmri" dataset using sns.load_dataset("fmri") and store it in the fmri_data variable.

3.We create a line plot using sns.lineplot() with "timepoint" on the x-axis, "signal" on the y-axis, and we use the hue parameter to color the lines based on the "event" column and the style parameter to distinguish between different "region" values.

4.We set labels for the x-axis and y-axis, add a title to the plot, and display the legend to differentiate the events.

5.Finally, we use plt.show() to display the plot.

This code will generate a line plot that shows how the "signal" variable changes over different "timepoints" for different "events" and "regions" in the "fmri" dataset.

Ans - To load the "titanic" dataset and create two box plots using Seaborn, one for 'age' and another for 'fare', with 'pclass' on the x-axis, you can follow these steps:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Create the first box plot for 'age'
sns.boxplot(ax=axes[0], data=titanic_data, x='pclass', y='age')
axes[0].set_title('Age by Pclass')
axes[0].set_xlabel('Pclass')
axes[0].set_ylabel('Age')

# Create the second box plot for 'fare'
sns.boxplot(ax=axes[1], data=titanic_data, x='pclass', y='fare')
axes[1].set_title('Fare by Pclass')
axes[1].set_xlabel('Pclass')
axes[1].set_ylabel('Fare')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```

3.We create a figure with two subplots using plt.subplots(1, 2) to have one row and two columns of plots.

4.In the first subplot (axes[0]), we create a box plot using sns.boxplot() for 'age' on the y-axis and 'pclass' on the x-axis. We set titles, x-axis labels, and y-axis labels for this subplot.

5.In the second subplot (axes[1]), we create another box plot for 'fare' on the y-axis and 'pclass' on the x-axis, and we set titles, x-axis labels, and y-axis labels for this subplot.

6.We adjust the spacing between subplots using plt.tight_layout().

7.Finally, we use plt.show() to display both box plots.

This code will generate two box plots, one showing the distribution of 'age' by 'pclass' and the other showing the distribution of 'fare' by 'pclass' in the "titanic" dataset.






Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

Ans - You can create a histogram for the 'price' column of the "diamonds" dataset from Seaborn and use the hue parameter to differentiate the data by the 'cut' column as follows:


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Create a histogram with hue for 'cut'
plt.figure(figsize=(10, 6))  # Optional: Set the figure size

# Use sns.histplot() with the 'price' column and 'cut' column as hue
sns.histplot(data=diamonds_data, x="price", hue="cut", bins=30, multiple="stack")

# Set plot labels and title
plt.xlabel("Price")
plt.ylabel("Count")
plt.title("Histogram of Diamond Prices by Cut")

# Show the legend
plt.legend(title="Cut")

# Show the plot
plt.show()
```

In this code:

1.We load the "diamonds" dataset using sns.load_dataset("diamonds") and store it in the diamonds_data variable.

2.We create a histogram using sns.histplot(). We specify the 'price' column as the data to plot on the x-axis and use the 'cut' column as the hue. The bins parameter determines the number of bins or intervals for the histogram, and multiple="stack" ensures that the data is stacked based on the 'cut' categories.

3.We set labels for the x-axis and y-axis, add a title to the plot, and display the legend to differentiate the 'cut' categories.

4.Finally, we use plt.show() to display the histogram with the 'cut' differentiation.

This code will generate a histogram of diamond prices, and each bar in the histogram will be stacked by the 'cut' category, allowing you to visualize the distribution of prices for different diamond cuts.






Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

Ans - You can create a pair plot for the "iris" dataset from Seaborn and use the hue parameter to differentiate the data by the "species" column as follows:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Create a pair plot with hue for "species"
sns.set(style="ticks")
sns.pairplot(iris_data, hue="species", markers=["o", "s", "D"])

# Show the plot
plt.show()
```

In this code:

1.We load the "iris" dataset using sns.load_dataset("iris") and store it in the iris_data variable.

2.We set the Seaborn style to "ticks" using sns.set(style="ticks") to provide a cleaner look to the pair plot.

3.We create the pair plot using sns.pairplot(). We specify the dataset (iris_data) and use the hue parameter to differentiate the data points by the "species" column. We also specify markers to distinguish the different species.

4.Finally, we use plt.show() to display the pair plot.

This code will generate a pair plot that shows scatterplots for all pairs of numerical features in the "iris" dataset. The data points will be colored and marked differently based on the "species" column, allowing you to visualize the relationships between variables while considering the different iris species.

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

Ans - Certainly! To plot a heatmap using the "flights" dataset from Seaborn, you can use the following code:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the dataset to create a matrix suitable for a heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
plt.figure(figsize=(10, 6))  # Optional: Set the figure size

# Use sns.heatmap() to create the heatmap
sns.heatmap(flights_pivot, cmap='coolwarm', annot=True, fmt='d', linewidths=0.5)

# Set plot labels and title
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Passenger Count Heatmap (Flights Dataset)")

# Show the plot
plt.show()
```

In this code:

1.We load the "flights" dataset using sns.load_dataset("flights") and store it in the flights_data variable.

2.To create a heatmap, we pivot the dataset using the pivot_table method, which restructures the data into a matrix format suitable for a heatmap. This step rearranges the data with months on the y-axis, years on the x-axis, and passenger counts as values.

3.We create the heatmap using sns.heatmap(). We specify the dataset (flights_pivot), colormap (cmap), annotate the cells with passenger counts using annot=True, format the annotations as integers using fmt='d', and set the linewidths between cells using linewidths=0.5.

4.We set labels for the x-axis and y-axis and add a title to the heatmap.

5.Finally, we use plt.show() to display the heatmap.

This code will generate a heatmap that visualizes the passenger counts over the years and months in the "flights" dataset. The color intensity represents the number of passengers for each month and year combination.
